```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.applications.xception import preprocess_input
```

```python
import numpy as np
```

```python
image = load_img('pants.jpg',target_size=(299, 299))

x = np.array(image)
X = np.array([x])

X = preprocess_input(X)
```

## Convert Keras to TF Lite

```python
model = keras.models.load_model('xception_v4_numpy_v2_28_0.894.keras')

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open('xception_v4_numpy_v2_28_0.894.tflite', 'wb') as f_out:
    f_out.write(tflite_model)
```

```python
import tensorflow.lite as tflite
```

```python
interpreter = tflite.Interpreter(model_path='xception_v4_numpy_v2_28_0.894.tflite')
interpreter.allocate_tensors()
```

```python
interpreter.get_input_details()
```

```python
interpreter.get_output_details()
```

```python
input_index = interpreter.get_input_details()[0]['index']
```

```python
ouput_index = interpreter.get_output_details()[0]['index']
```

```python
interpreter.set_tensor(input_index, X)
interpreter.invoke()
preds = interpreter.get_tensor(ouput_index)
```

```python
classes = [
    'dress',
    'hat',
    'longsleeve',
    'outwear',
    'pants',
    'shirt',
    'shoes',
    'shorts',
    'skirt',
    't-shirt'
]

dict(zip(classes, preds[0]))
```

## Removing tensorflow dependency

```python
from PIL import Image
```

```python
with Image.open('pants.jpg') as img:
    img = img.resize((299, 299), Image.NEAREST)
```

```python
def preprocess_input(x):
    x /= 127.5
    x -= 1.
    return x
```

```python
x = np.array(image, dtype='float32')
X = np.array([x])

X = preprocess_input(X)
```

```python
interpreter.set_tensor(input_index, X)
interpreter.invoke()
preds = interpreter.get_tensor(ouput_index)
```

```python
classes = [
    'dress',
    'hat',
    'longsleeve',
    'outwear',
    'pants',
    'shirt',
    'shoes',
    'shorts',
    'skirt',
    't-shirt'
]

dict(zip(classes, preds[0]))
```

```python
from keras_image_helper import create_preprocessor
```

```python
preprocessor = create_preprocessor('xception', target_size=(299, 299))
```

```python
url = 'http://bit.ly/mlbookcamp-pants'
X = preprocessor.from_url(url)
```

```python
interpreter.set_tensor(input_index, X)
interpreter.invoke()
preds = interpreter.get_tensor(ouput_index)
```
